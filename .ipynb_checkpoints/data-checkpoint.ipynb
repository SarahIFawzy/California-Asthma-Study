```python
# link to dataset 1: CalEnviroScreen Final Report https://data.ca.gov/dataset/calenviroscreen-2-0/resource/6d6a9a5c-9221-4fc5-9b00-35015ae65837
import json
import urllib.request
import pandas as pd

# Define the URL for the API request without the query filter
url = 'https://data.ca.gov/api/3/action/datastore_search?resource_id=6d6a9a5c-9221-4fc5-9b00-35015ae65837&limit=20'

# Fetch the data from the URL
fileobj = urllib.request.urlopen(url)
response_dict = json.loads(fileobj.read())

# Print the JSON response to inspect its structure
print(json.dumps(response_dict, indent=4))

# Extract the records from the response
records = response_dict.get('result', {}).get('records', [])

# Check if records are found
if not records:
    print("No records found in the JSON response.")
else:
    # Convert the records to a DataFrame
    survey_1_df = pd.DataFrame(records)

    # Display the first 20 rows
    survey_1_df.head()
```
